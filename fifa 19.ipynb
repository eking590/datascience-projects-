# weights/criteria for analysis 

```python
a = 0.5 
b = 1 
c= 2 
d = 3 
```

# best goal keeper 

```python
df['goalkeeper_stopper'] = (b*df.ShortPassing + a* df.Volleys + c * df.LongPassing + b * df.BallControl + d * df.Jumping + b * df.Composure + d * df.GKDiving + d * df.GKHandling + d * df.GKKicking + c * df.Positioning + d * df.GKReflexes + a * df.Reactions + c * df.Agility)/(2*a+4*b+2*c+1*d) 
df['goal_sweepers'] = (a*df.Reactions + b*df.ShortPassing + a*df.Volleys + c * df.LongPassing + b *df.BallControl + c*df.Jumping + b*df.Composure + d*df.GKDiving + d*df.GKHandling + d*df.GKKicking + c* df.Positioning + d * df.GKReflexes + b*df.Agility)/ (2*a+4*b+2*c+1*d)
```

```python
plt.figure(figsize=(15,6)) 

sd1 = df.sort_values('goalkeeper_stopper', ascending=False)[:10] 
x1 = np.array(list(sd1['Name'])) 
y2 = np.array(list(sd1['goalkeeper_stopper'])) 

sns.barplot(x1, y2, palette='colorblind') 
plt.ylabel('Shot Stopping Score')
```

```python
plt.figure(figsize=(15,6)) 

sd1 = df.sort_values('goal_sweepers', ascending=False)[:10] 
x1 = np.array(list(sd1['Name'])) 
y2 = np.array(list(sd1['goal_sweepers'])) 

sns.barplot(x1, y2, palette='colorblind') 
plt.ylabel('goal sweeper score')
```
